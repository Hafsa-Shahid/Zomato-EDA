### ZOMATO - ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Zomato - Ready to Process.csv")
```

### Popular Cuisines: 
Analyzing the 'cuisines' column to identify the most popular cuisines in different cities or localities.

This plot quickly identifies which cuisines are more prevalent in the restaurants. 

```python
plt.figure(figsize=(12, 6))

top_cuisines = df['cuisines'].value_counts().head(10)

sb.barplot(x = top_cuisines.index, y = top_cuisines.values, palette = 'viridis')
plt.title('Top 10 Popular Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation = 45)
plt.show()
```

### Average Cost: 
Exploring the distribution of the 'average cost for two' to understand the cost range for dining in different areas.

This histogram provides insights into the distribution of average costs, helping in understanding the affordability range for dining in different areas. This key info can guide marketing campaigns, such as promotions targeting specific price ranges, loyalty programs, or special offers for different customer segments.Here x-axis represents the average cost for two people at a restaurant and y-axis represents the frequency of restaurants within each bin.

```python
plt.figure(figsize=(12, 6))

sb.histplot(df['average_cost_for_two'], bins = 50, color = 'blue', kde = True)

plt.title('Distribution of Average Cost for Two')
plt.xlabel('Average Cost for Two')
plt.ylabel('Count of Restaurants')
plt.show()
```

### Price Range:
Investigating the distribution of the 'price range' to understand the affordability of restaurants in different regions.

For exploring the distribution of the 'price range' column, we use a count plot to visualize the occurance of different
price ranges. This can help us understand the affordability categories of the restaurants in the dataset.

In this plot, x-axis represents the different price categories of restaurants and y-axis represents the number of restaurants in each price range.
This count plot allows us to see how many restaurants fall into each price range category and can be valuable for marketing strategies by targeting specific price segments, understanding customer preferences and tailoring offerings to the market.

```python
plt.figure(figsize=(10, 6))

sb.countplot(x = 'price_range', data = df, palette = 'Set2')
plt.title('Distribution of Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.show()
```

### Ratings and Votes:

Analyzing the 'aggregate rating', 'rating text', and 'votes' columns to identify top-rated restaurants and those with high customer engagement.
To explore the relationship between ratings and votes, a scatter plot can be informative. This plot will help in visualizing how the aggregate rating relates to the number of votes a restaurant has received.

In this plot, x-axis represents the aggregate rating of each restaurant and y-axis represents the number of votes received by each restaurant. This scatter plot allows us to visually assess if there is any correlation between the aggregate rating and the number of votes. Understanding this relationship is crucial for planning marketing campaigns that encourage customers to leave reviews and engage with the restaurant.

```python
plt.figure(figsize=(10, 6))

sb.scatterplot (x = 'aggregate_rating', y = 'votes', data = df, color = 'coral')

plt.title('Relationship between Ratings and Votes')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Votes')
plt.show()
```

### Establishment Type:

Explore the 'establishment' column to understand the types of dining establishments in different cities.

In this plot, the x-axis represents different types of dining establishments and y-axis represents the count of restaurants for each establishment type.
This plot visualizes the distribution of different dining establishments, helping in understanding the landscape of restaurant offerings in the dataset.

```python
plt.figure(figsize=(14, 6))

sb.countplot(x = 'establishment', data = df, palette = 'pastel', order = df['establishment'].value_counts().index)

plt.title('Distribution of Dining Establishments')
plt.xlabel('Establishment Type')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation = 45, ha = 'right') 
plt.show()
```

### City - Wise Analysis:

Comparison and contrast of dining trends and preferences in different cities.

```python
plt.figure(figsize=(22, 10))

sb.countplot(x = 'city', data = df, palette = 'Set3', order = df['city'].value_counts().index)

plt.title('Distribution of Restaurants Across Cities')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation = 90, ha='right')  
plt.show()
```

### Correlation Analysis

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numerical_columns].corr()

plt.figure(figsize=(8, 8))
sb.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

### HISTOGRAM PLOTS:
 we can see that non of the feature is normally distributed.


```python
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
# Plotting distributions of various features
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(20, 20))

columns = df.columns
index = 0

for i in range(3):
    for j in range(2):
        if index < len(numeric_columns):
            sb.histplot(df[numeric_columns[index]], kde=True, ax=axes[i, j])
            axes[i, j].set_title(numeric_columns[index])
            index += 1

plt.tight_layout()
plt.show()
```

### BOX PLOTS FOR OUTLIERS

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Boxplots to visualize outliers in numerical columns
plt.figure(figsize=(15, 10))

# Creating a boxplot for each numerical feature in the DataFrame
for i, column in enumerate(numeric_columns):
    plt.subplot(4, 4, i + 1)
    sb.boxplot(y=df[column])
    plt.title(column)
    plt.tight_layout()

plt.show()
```
