```python
df = df.drop(columns = ['opentable_support'])
df = df.drop(columns = ['zipcode'])
df = df.drop(columns = ['delivery'])
df = df.drop(columns = ['takeaway'])
df = df.drop(columns = ['url'])
df = df.drop(columns = ['address'])
df = df.drop(columns = ['timings'])
```

```python
df = df.drop(columns = ['country_id'])
df = df.drop(columns = ['locality_verbose'])
df = df.drop(columns = ['currency'])
df = df.drop(columns = ['photo_count'])
```

```python
df.info()
```

```python
df = df.drop(columns = ['latitude'])
df = df.drop(columns = ['longitude'])
```

```python
df['establishment'] = df['establishment'].str.replace('[]', 'unknown')
```

```python
df['establishment'] = df['establishment'].str.strip("[]'")
```

```python
df['cuisines'].isnull().sum()
```

```python
df['cuisines'] = df['cuisines'].fillna('Not Available')
```

```python
df['cuisines'].isnull().sum()
```

```python
df['highlights'] = df['highlights'].str.replace('[]', 'Not Available')
```

```python
df['highlights'] = df['highlights'].str.strip("[]")
```
